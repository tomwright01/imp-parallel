# Cone counting example

## Tasks:
* Load the image file data/sample.png as a numpy.nd array
* Apply a gaussian filter to the image
* Identify the regional maximas in the image
* Count the number of located maxima

## Suggested functions:
* scipy.misc.imread
* scipy.ndimage.gaussian_filter
* scipy.ndimage.filters.maximum_filter
* scipy.ndimage.label


```python
# setup the environment
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc
import scipy.ndimage
%matplotlib inline
```

```python
image = scipy.misc.imread("data/sample.png", mode="P")
image.shape
```

```python
plt.gray()
plt.imshow(image)
```

```python
image_filtered = scipy.ndimage.gaussian_filter(image,3)
```

```python
plt.imshow(image_filtered)
```

```python
image_maxima = scipy.ndimage.filters.maximum_filter(image_filtered, size=5)
```

```python
plt.imshow(image_maxima)
```

```python
image_maxima[0:10,0:10]
```

```python
cones_masked = image_filtered == image_maxima
```

```python
cones_masked.dtype
```

```python
labels = scipy.ndimage.label(cones_masked)
```

```python
print("There are {} cones in this image".format(labels[1]))
```
